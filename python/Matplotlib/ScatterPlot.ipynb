```python
plt.scatter(df['avg'],df['strike_rate'] , color='red',marker='.',s=df['runs']/10)
plt.title("Average vs Strike Rate")
plt.xlabel("Average")
plt.ylabel("Strike Rate")
plt.grid(True)
```

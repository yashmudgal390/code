```python
plt.plot(df['index'],df['V Kohli'])
plt.plot(df['index'],df['RG Sharma'])
plt.title("V Kohli vs RG Sharma")
plt.xlabel("Season")
plt.ylabel("Runs")
```

```python
plt.plot(df['index'],df['V Kohli'],color='red')
plt.plot(df['index'],df['RG Sharma'],color='blue')
plt.title("V Kohli vs RG Sharma")
plt.xlabel("Season")
plt.ylabel("Runs")
```

```python
#color style marker makersze line width lable 
plt.plot(df['index'],df['V Kohli'],color='red',linestyle='dashdot',linewidth=2,label='V Kohli',marker='o',markersize=4)
plt.plot(df['index'],df['RG Sharma'],color='blue',linestyle='dashdot',linewidth=2,label='RG Sharma',marker='o',markersize=8)
plt.title("V Kohli vs RG Sharma")
plt.xlabel("Season")
plt.ylabel("Runs")
plt.grid()

plt.legend()
```

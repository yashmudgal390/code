```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
children = [10,20,30,40,10]
colors = ['red','blue','green','yellow','cyan']
plt.bar(colors,children)
```

```python
children = [10,20,30,40,10]
colors = ['red','blue','green','yellow','cyan']
plt.barh(colors,children)
```

```python
df = pd.read_csv('batsman_season_record.csv')
```

```python
df
```

```python
plt.bar(np.arange(df.shape[0]) - 0.3 ,df['2015'], width=0.2)
plt.bar(np.arange(df.shape[0])   ,df['2016'], width=0.2)
plt.bar(np.arange(df.shape[0]) + 0.3 ,df['2017'], width=0.2)
plt.xticks(np.arange(df.shape[0]), df['batsman'], rotation=45)
```

```python
#stacked bar chart
plt.bar(df['batsman'], df['2017'])
plt.bar(df['batsman'], df['2016'], bottom=df['2017'])
plt.bar(df['batsman'], df['2015'], bottom=df['2016']+df['2017'])
```

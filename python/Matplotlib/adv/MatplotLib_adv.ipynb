```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('iris.csv')
df.head()
```

```python
df['Species'] = df['Species'].replace({'Iris-setosa': 0, 'Iris-versicolor': 1, 'Iris-virginica': 2})
df.sample(5)
```

```python
plt.scatter(df['SepalLengthCm'],df['PetalLengthCm'],c=df['Species'],cmap='summer',alpha=0.7)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

```python
plt.figure(figsize=(15,7))
plt.scatter(df['SepalLengthCm'],df['PetalLengthCm'],c=df['Species'],alpha=0.7)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

```python
df = pd.read_csv('batter.csv')
df.head()
```

```python
df = df.head(100).sample(25,random_state=5)
df.head()
```

```python
plt.figure(figsize=(15,7))
plt.scatter(df['avg'],df['strike_rate'],s=df['runs'],alpha=0.7)
for i in range(df.shape[0]):
    plt.text(df['avg'].values[i],df['strike_rate'].values[i],df['batter'].values[i])
```

```python
plt.figure(figsize=(18,13))
plt.scatter(df['avg'],df['strike_rate'],s=df['runs'],alpha=0.7)

plt.axhline(130,color='red',linestyle='-')
plt.axhline(140,color='green',linestyle='-')
plt.axvline(30,color='red',linestyle='-')
plt.xlabel('Average')
plt.ylabel('Strike Rate')
for i in range(df.shape[0]):
    plt.text(df['avg'].values[i],df['strike_rate'].values[i],df['batter'].values[i])
```

```python
fig , ax = plt.subplots(figsize=(18,13))

ax.scatter(df['avg'],df['strike_rate'],s=df['runs'],alpha=0.7)
ax.set_title('Batter Analysis')
ax.set_xlabel('Average')
ax.set_ylabel('Strike Rate')
```

```python
fig , ax = plt.subplots(nrows=2,ncols=2,figsize=(10,10),sharex= True)  
```

```python
fig , ax = plt.subplots(nrows=2,ncols=2,figsize=(10,10))  
ax[0,0].scatter(df['avg'],df['strike_rate'])
ax[0,1].scatter(df['avg'],df['runs'])
ax[1,0].hist(df['avg'])
ax[1,1].hist(df['runs'])
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)
ax3 = fig.add_subplot(223)

ax1.scatter(df['avg'],df['strike_rate'],color='red')
ax2.scatter(df['avg'],df['runs'],color='green')
ax3.hist(df['avg'])
```

<h3>3d scatter plot</h3>

```python
df = pd.read_csv('batter.csv')
df
```

```python
fig = plt.figure()
ax = plt.subplot(projection='3d')
ax.scatter3D(df['runs'], df['avg'], df['strike_rate'])
ax.set_title('3D Scatter Plot')
ax.set_xlabel('Runs')
ax.set_ylabel('Average')
ax.set_zlabel('Strike Rate')
plt.show()
```

```python
x=[0,1,5]
y = [0,19,13]
z= [0,13, 20]
fig = plt.figure()
ax = plt.subplot(projection='3d')
ax.scatter3D(x, y, z)
ax.plot3D(x, y, z)
```

```python
#3d surface plot 
x = np.linspace(-10,10, 100)
y = np.linspace(-10,10, 100)
x
```

```python
xx, yy = np.meshgrid(x, y)
```

```python
xx.shape
```

```python
yy.shape
```

```python
z= xx**2+ yy**2
z.shape
```

```python
fig=plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d')
p=ax.plot_surface(xx, yy, z, cmap='viridis')
fig.colorbar(p)
```

```python
z=np.sin(xx)+np.cos(yy)
fig=plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d')
p=ax.plot_surface(xx, yy, z, cmap='viridis')
fig.colorbar(p)
```

```python
fig=plt.figure(figsize=(12,8))
ax = plt.subplot()
p=ax.contour(xx, yy, z, cmap='viridis')
fig.colorbar(p)
```

```python
fig=plt.figure(figsize=(12,8))
ax = plt.subplot()
p=ax.contourf(xx, yy, z, cmap='viridis')
fig.colorbar(p)
```

```python
z=np.sin(xx)+np.cos(yy)
fig=plt.figure(figsize=(12,8))
ax = plt.subplot()
p=ax.contourf(xx, yy, z, cmap='viridis')
fig.colorbar(p)
```

```python
delivery = pd.read_csv('IPL_Ball_by_Ball_2008_2022.csv')
delivery.head()
```

```python
temp_df=delivery[(delivery['ballnumber'].isin([1,2,3,4,5,6])) & (delivery['batsman_run']==6)]
```

```python
grid = temp_df.pivot_table(index='overs',columns='ballnumber',values='batsman_run',aggfunc='count')
```

```python
plt.figure(figsize=(12,8))
plt.imshow(grid,cmap='viridis')
plt.yticks(delivery['overs'].unique(),list(range(1,21)))
plt.xticks(np.arange(0,6),list(range(1,7)))

plt.colorbar()
```

```python
import numpy as np 
import pandas as pd
```

```python
employees = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],
    'name': ['John', 'Anna', 'Peter', 'Linda', 'Bob'],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR']
})

# DataFrame 2: Salary information
salaries = pd.DataFrame({
    'employee_id': [1, 2, 3, 6, 7],
    'salary': [60000, 80000, 65000, 70000, 90000],
    'bonus': [5000, 10000, 7000, 8000, 12000]
})
employees
salaries
```

```python
employees.merge(salaries,on='employee_id',how='inner')
```

```python
employees.merge(salaries, on='employee_id', how='outer')
```

```python
employees.merge(salaries, on='employee_id', how='left')
```

```python
employees.merge(salaries, on='employee_id', how='right')
```

```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('imdb-top-1000.csv')
movies
```

```python
gen = movies.groupby('Genre')
gen
```

```python
gen.sum()['Gross'].sort_values(ascending=False).head(3)
```

```python
df = pd.DataFrame(columns=movies.columns)
for group,data in gen:
    df = pd.concat([df, data[data['IMDB_Rating'] == data['IMDB_Rating'].max()]], ignore_index=True)
df
```

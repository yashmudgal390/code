```python
movies = pd.read_csv('imdb-top-1000.csv')
movies
```

```python
gen = movies.groupby('Genre')
gen
```

```python
gen.sum()['Gross'].sort_values(ascending=False).head(3)
```

```python
df = pd.DataFrame(columns=movies.columns)
for group,data in gen:
    df = pd.concat([df, data[data['IMDB_Rating'] == data['IMDB_Rating'].max()]])
df
```

```python
def fo(group):
    return group['Series_Title'].str.startswith('A').sum()
```

```python
gen.apply(fo)
```

```python
def rank(group):
    group['Rank']=group['IMDB_Rating'].rank(ascending=False)
    return group
```

```python
gen.apply(rank)
```

```python
def normal(grp):
    grp['nor_rating']=(grp['IMDB_Rating']-grp['IMDB_Rating'].min())/(grp['IMDB_Rating'].max()-grp['IMDB_Rating'].min())
    return grp
```

```python
gen.apply(normal)
```

```python
def max_runs_per_player(df):
    player_match_runs = df.groupby(['batsman', 'match_id'])['batsman_runs'].sum().reset_index()
    
    max_runs = player_match_runs.groupby('batsman')['batsman_runs'].max().reset_index()
    
    max_runs.rename(columns={'batsman_runs': 'max_runs'}, inplace=True)
    
    
    return max_runs
```

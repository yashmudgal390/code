```python
def head_to_head(ipl, team1, team2, detailed=False):
    # Filter matches between the two teams
    matches = ipl[
        ((ipl['Team1'] == team1) & (ipl['Team2'] == team2)) |
        ((ipl['Team1'] == team2) & (ipl['Team2'] == team1))
    ]
    
    total_matches = len(matches)
    
    # Wins count
    team1_wins = len(matches[matches['WinningTeam'] == team1])
    team2_wins = len(matches[matches['WinningTeam'] == team2])
    
    # No result / tie
    no_result = total_matches - (team1_wins + team2_wins)
    
    print(f"\n📊 {team1} vs {team2}")
    print(f"Total Matches: {total_matches}")
    print(f"{team1} Wins: {team1_wins}")
    print(f"{team2} Wins: {team2_wins}")
    print(f"No Result/Tie: {no_result}")
    
    if detailed:
        return matches[['Date', 'Season', 'Team1', 'Team2', 'WinningTeam', 'Margin', 'Player_of_Match']]
```

```python
head_to_head(ipl, 'Mumbai Indians', 'Chennai Super Kings')
```

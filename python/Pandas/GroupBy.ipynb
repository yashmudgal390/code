```python
import numpy as np 
import pandas as pd
```

```python
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Store': ['S1', 'S1', 'S2', 'S2', 'S1', 'S2', 'S2', 'S1'],
    'Sales': [100, 200, 150, 250, 120, 180, 200, 300],
    'Quantity': [10, 15, 12, 18, 8, 20, 15, 25],
    'Date': pd.date_range('2023-01-01', periods=8)
}
df = pd.DataFrame(data)
df
```

```python
gr = df.groupby('Store')['Sales'].sum()
gr
```

```python
gr = df.groupby(['Store', 'Category'])['Sales'].sum()
gr
```

```python
gr = df.groupby([ 'Category','Store']).agg({'Sales':['sum','mean','std','min','max','size','count', 'nunique', 'first', 'last']})
gr
```
